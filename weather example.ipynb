```python
import requests
import pandas as pd
from pprint import pprint as pp
```

```python
json_city = requests.get("https://www.metaweather.com/api/location/search/?query=Miami").json()
pp(json_city)
```

```python
id = json_city[0]["woeid"]
print(id)
```

```python
pp(requests.get("https://www.metaweather.com/api/location/"+str(id)+"/2019/10/11/").json()[:30])
```
